```python
import requests 
import pandas as pd


#searching for the tesla symbol
api_key = 'TOXB8F6GQ0AHUYOA'
key = 'tesla'
url = 'https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords=tesla&apikey=api_key'
r = requests.get(url)
data = r.json()

print(data)
```

```python
#Fetching the intraday data of the tesla stock

symbol = 'TSLA'
interval = "5min"
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=TSLA&interval={interval}&outputsize=full&apikey=TOXB8F6GQ0AHUYOA"
r = requests.get(url)
data = r.json()

print(data)
```

```python
#Converting the json data to pandas dataframe

time_series = data[f"Time Series ({interval})"]

df = pd.DataFrame.from_dict(time_series, orient="index")
df.columns = ["Open", "High", "Low", "Close", "Volume"]
df.index = pd.to_datetime(df.index)
df = df.astype(float)
print(df)
```

```python
#Fetching the RSI values from the API
function = "RSI"
url = 'https://www.alphavantage.co/query?function=RSI&symbol=TSLA&interval=5min&time_period=10&series_type=open&apikey=api_key'
r = requests.get(url)
data_rsi = r.json()

print(data_rsi)
```

```python
#Converting the json data to pandas dataframe

time_series = data_rsi[f"Technical Analysis: {function}"]

rsi_df = pd.DataFrame.from_dict(time_series,orient="index")
rsi_df.columns = ["RSI"]
rsi_df.index = pd.to_datetime(rsi_df.index)
rsi_df = rsi_df.astype(float)
print(rsi_df)
```

```python
#merged the data obtained from the intraday and the RSI 

merged_df = pd.merge(df, rsi_df, left_index=True, right_index=True, how='outer')

print(merged_df)
```

```python
#dropping the timestamps for which I don't have the RSI value

merged_df = merged_df.dropna(subset=['RSI'])
print(merged_df)
```

```python
#Trading strategy I used is quite simple which is buy when the stock is oversold which is when RSI is less than 30 and buy the stock when the stock is overbought which is when the RSI is greater than 70

merged_df['Buy_Signal'] = merged_df['RSI'] < 30
merged_df['Sell_Signal'] = merged_df['RSI'] > 70

initial_capital = 100000
shares = 0

capital = initial_capital

merged_df['Portfolio_Value'] = float(capital)

def calculate_returns(merged_df, capital, shares):
    
    for i in range(1, len(merged_df)):
        if merged_df.iloc[i]['Buy_Signal'] and not merged_df.iloc[i-1]['Buy_Signal']:
            shares_to_buy = capital // merged_df.iloc[i]['Close']
            shares += shares_to_buy
            capital -= shares_to_buy * merged_df.iloc[i]['Close']

        elif merged_df.iloc[i]['Sell_Signal'] and not merged_df.iloc[i-1]['Sell_Signal'] and shares > 0:
            capital += shares * merged_df.iloc[i]['Close']
            shares = 0

        merged_df.at[merged_df.index[i],'Portfolio_Value'] = capital + shares * merged_df.iloc[i]['Close']
    
    final_portfolio_value = merged_df['Portfolio_Value'].iloc[-1]
    return final_portfolio_value

returns = calculate_returns(merged_df, capital, shares)

#printing the returns
print(returns)
    


```

```python
#Let's prepare the data for the year 2023 to test our strategy performance

symbol = 'TSLA'
interval = "5min"
year = "2023"

combined_df = pd.DataFrame()

for date in range(1, 13):
    month = f"{date:02d}"
   
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval={interval}&month={year}-{month}&outputsize=full&apikey=TOXB8F6GQ0AHUYOA"
    r = requests.get(url)
    data = r.json()
    time_series = data[f"Time Series ({interval})"]

    df = pd.DataFrame.from_dict(time_series, orient="index")
    df.columns = ["Open", "High", "Low", "Close", "Volume"]
    df.index = pd.to_datetime(df.index)
    df = df.astype(float)
    
    combined_df = pd.concat([combined_df, df])

    
print(combined_df)
```

```python
#FEtching RSI data for the year 2023
function = "RSI"
timePeriod="10"

combined_rsi_df = pd.DataFrame()

for date in range(1, 13):
    month = f"{date:02d}"
    url = f"https://www.alphavantage.co/query?function={function}&symbol={symbol}&interval={interval}&time_period={timePeriod}&month={year}-{month}&series_type=open&apikey=api_key"
    r = requests.get(url)
    data_rsi = r.json()
    time_series = data_rsi[f"Technical Analysis: {function}"]

    rsi_df = pd.DataFrame.from_dict(time_series,orient="index")
    rsi_df.columns = ["RSI"]
    rsi_df.index = pd.to_datetime(rsi_df.index)
    rsi_df = rsi_df.astype(float)
    
    combined_rsi_df = pd.concat([combined_rsi_df, rsi_df])
    
print(combined_rsi_df)
```

```python
#Merging the data

merged_2023_df = pd.merge(combined_df, combined_rsi_df, left_index=True, right_index=True, how='outer')
merged_2023_df = merged_2023_df.dropna(subset=['RSI'])
print(merged_2023_df)
```
